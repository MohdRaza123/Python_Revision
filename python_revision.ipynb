**Task 1: NumPy Basics**

Write a Python script that creates a NumPy array with random integers and performs basic mathematical operations like addition, subtraction, multiplication, and division on it.

```python
import numpy as np

# Create a NumPy array with random integers
array_size = 5  # You can change this to the desired size of the array
min_value = 1   # Minimum value for random integers
max_value = 100 # Maximum value for random integers
random_array = np.random.randint(min_value, max_value, size=array_size)

# Display the random array
print("Random Array:", random_array)

# Perform basic mathematical operations
scalar = 5

# Addition
addition_result = random_array + scalar
print("Addition Result:", addition_result)

# Subtraction
subtraction_result = random_array - scalar
print("Subtraction Result:", subtraction_result)

# Multiplication
multiplication_result = random_array * scalar
print("Multiplication Result:", multiplication_result)

# Division (Make sure to handle division by zero)
divisor = 0
division_result = np.divide(random_array, divisor, where=divisor!=0)
print("Division Result:", division_result)
```

**Task 2: Pandas Data Analysis**

Using Pandas, load a dataset of your choice (you can find datasets online or use your own). Clean the data and perform basic data analysis tasks like calculating mean, median, and mode for specific columns.


```python
import pandas as pd

# Load the dataset with the specified encoding
file_path = "../content/drive/MyDrive/Online Retail.csv"
df = pd.read_csv(file_path, encoding='latin1')  # Try 'latin1' encoding

# Data Cleaning (You can customize this step as needed)
# For this example, let's assume you want to remove rows with missing values.
df = df.dropna()

# Data Analysis
# Calculate mean, median, and mode for specific columns Quantity
target_column = "Quantity"  # Replace with the column you want to analyze

if target_column in df.columns:
    # Calculate the mean
    mean_value = df[target_column].mean()
    print(f"Mean of {target_column}: {mean_value}")

    # Calculate the median
    median_value = df[target_column].median()
    print(f"Median of {target_column}: {median_value}")

    # Calculate the mode (Note: There can be multiple modes)
    mode_values = df[target_column].mode()
    print(f"Mode(s) of {target_column}: {', '.join(map(str, mode_values))}")
else:
    print(f"'{target_column}' not found in the dataset.")

# You can repeat the above code for other columns as needed.
```

**Task 4: NumPy Array Manipulation**

Create two NumPy arrays and perform operations to concatenate them vertically and horizontally.


```python
import numpy as np

# Create two NumPy arrays
array1 = np.array([[1, 2, 3],
                   [4, 5, 6]])

array2 = np.array([[7, 8, 9],
                   [10, 11, 12]])

# Vertically stack (concatenate along rows)
vertical_stack = np.vstack((array1, array2))

# Horizontally stack (concatenate along columns)
horizontal_stack = np.hstack((array1, array2))

# Display the original arrays and the concatenated results
print("Array 1:")
print(array1)

print("Array 2:")
print(array2)

print("Vertically Stacked:")
print(vertical_stack)

print("Horizontally Stacked:")
print(horizontal_stack)
```

**Task 5: Pandas Data Filtering**

Given a dataset (you can choose one or use your own), use Pandas to filter and extract rows that meet specific criteria (e.g., filtering data for a specific date range).


```python
import pandas as pd

# Load the dataset with the specified encoding
file_path = "../content/drive/MyDrive/Online Retail.csv"
df = pd.read_csv(file_path, encoding='latin1')  # Try 'latin1' encoding

# Assuming your dataset has a column named 'date_column_name' which contains date values
date_column_name = "InvoiceDate"  #  the actual date column name

# Define the date range you want to filter for
start_date = "01/12/2010  08:34:00"  #  desired start date
end_date = "31/10/2011  14:41:00"    #  desired end date

# Convert the date column to datetime format (if it's not already)
df[date_column_name] = pd.to_datetime(df[date_column_name])

# Filter the dataset for the specified date range
filtered_df = df[(df[date_column_name] >= start_date) & (df[date_column_name] <= end_date)]

# Display the filtered data
print("Filtered Data:")
print(filtered_df)
```

**Task 7: NumPy Statistical Analysis**

Use NumPy to calculate descriptive statistics like variance, standard deviation, and correlation coefficients for a dataset of your choice (you can find datasets online or use your own).


```python
import numpy as np

# Create or load your dataset
# For this example, let's create a sample dataset
data = np.array([[1, 2, 3, 4, 5],
                 [5, 4, 3, 2, 1],
                 [2, 3, 2, 4, 3]])

# Calculate the variance and standard deviation
variance = np.var(data, axis=1)  # Variance along each row
std_deviation = np.std(data, axis=1)  # Standard deviation along each row

# Calculate the correlation coefficient
correlation_matrix = np.corrcoef(data)
correlation_coefficient = correlation_matrix[0, 1]  # Assuming you want the coefficient between the first two rows

# Display the results
print("Variance for each row:", variance)
print("Standard Deviation for each row:", std_deviation)
print(f"Correlation Coefficient between the first two rows: {correlation_coefficient:.2f}")
```

**Task 8: Pandas Data Grouping**

Given a dataset (you can choose one or use your own), use Pandas to group data by a specific column and calculate summary statistics (e.g., mean, median) for each group.


```python
import pandas as pd

# Load the dataset with the specified encoding
file_path = "../content/drive/MyDrive/Online Retail.csv"
df = pd.read_csv(file_path, encoding='latin1')  # Try 'latin1' encoding

# Assuming your dataset has a 'group_column_name' for grouping and a 'value_column_name' for calculations
group_column_name = "Country"  #  the column to group by
value_column_name = "Quantity"  # column for which you want to calculate summary statistics

# Group the data by the 'group_column_name'
grouped = df.groupby(group_column_name)

# Calculate summary statistics (mean and median) for each group
summary_stats = grouped[value_column_name].agg(['mean', 'median'])

# Display the summary statistics
print("Summary Statistics:")
print(summary_stats)
```

**Task 9: Matplotlib Subplots**

Create a Matplotlib figure with multiple subplots to display different aspects of a dataset or related datasets (you can choose the dataset and the type of plots).


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset from Seaborn
data = sns.load_dataset("iris")

# Create a Matplotlib figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 grid of subplots

# Plot 1: Scatter plot of sepal length vs. sepal width
sns.scatterplot(x="sepal_length", y="sepal_width", data=data, ax=axes[0, 0])
axes[0, 0].set_title("Sepal Length vs. Sepal Width")

# Plot 2: Histogram of petal length
sns.histplot(data["petal_length"], ax=axes[0, 1], kde=True, color="green")
axes[0, 1].set_title("Petal Length Histogram")

# Plot 3: Box plot of petal width by species
sns.boxplot(x="species", y="petal_width", data=data, ax=axes[1, 0])
axes[1, 0].set_title("Petal Width by Species")

# Plot 4: Violin plot of sepal length by species
sns.violinplot(x="species", y="sepal_length", data=data, ax=axes[1, 1])
axes[1, 1].set_title("Sepal Length by Species")

# Adjust layout
plt.tight_layout()

# Show the figure with subplots
plt.show()
```

**Task 10: Pandas Data Visualization**

Choose a dataset of your choice and create various visualizations (e.g., bar charts, pie charts, scatter plots) using Pandas to represent different aspects of the data, such as trends or distributions.


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load a sample dataset from Seaborn
data = sns.load_dataset("tips")

# Visualize different aspects of the dataset using Pandas

# Bar chart: Count of total bills per day
total_bills_per_day = data.groupby("day")["total_bill"].sum()
total_bills_per_day.plot(kind="bar", color="skyblue")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount")
plt.title("Total Bills per Day")
plt.show()

# Pie chart: Proportion of meals by time (lunch or dinner)
meal_time_counts = data["time"].value_counts()
meal_time_counts.plot(kind="pie", autopct='%1.1f%%', colors=["lightcoral", "lightgreen"])
plt.title("Proportion of Meals by Time")
plt.axis("equal")  # Equal aspect ratio ensures a circular pie chart
plt.show()

# Scatter plot: Relationship between total bill and tip
data.plot.scatter(x="total_bill", y="tip", color="purple", marker="o")
plt.xlabel("Total Bill Amount")
plt.ylabel("Tip Amount")
plt.title("Total Bill vs. Tip Amount")
plt.show()
```
